```r
listGEOSuppFiles <- function(accession) {
  # Get the supplementary files without downloading them
  files_info <- getGEOSuppFiles(accession, fetch_files = FALSE)
  
  # Extract file names and URLs
  files <- data.frame(
    file_name = basename(rownames(files_info)),
    url = files_info$url,
    stringsAsFactors = FALSE
  )
  
  return(files)
}

# Function to list all primary files for a given GEO accession
listGEOFiles <- function(accession) {
  # Get the primary files without downloading them
  file_info <- getGEOfile(accession)
  
  # Extract file names and URLs
  files <- data.frame(
    file_name = basename(file_info$localfile),
    url = file_info$url,
    stringsAsFactors = FALSE
  )
  
  return(files)
}

# Main function to list all files for a given GEO accession
listAllGEOFiles <- function(accession) {
  supp_files <- listGEOSuppFiles(accession)
  primary_files <- listGEOFiles(accession)
  
  all_files <- rbind(supp_files, primary_files)
  return(all_files)
}

# Example usage
accession <- "GSE178333" # Replace with your GEO accession
all_files <- listAllGEOFiles(accession)

# Print the files
print(all_files)
```
